```python
import requests

url = 'https://api.github.com'
token = '******'
user = '******'

response = requests.get(
url + "/search/repositories",
params={
	'q': 'GPT in:name language:Python followers:>20000',
	'sort': 'stars',
	'order': 'desc'
},
headers={
	'Authorization': f'Bearer {token}',
	'X-GitHub-Api-Version': '2022-11-28',
	'Accept': 'application/vnd.github+json',
	'User-Agent': user
}
)

response.text
```

```python
data=response.json()
repo=data['items'][0]['name']
name=data['items'][0]['owner']['login']

```

```python
response = requests.get(
url + f"/repos/{name}/{repo}/commits",
headers={
	'Authorization': f'Bearer {token}',
	'X-GitHub-Api-Version': '2022-11-28',
	'Accept': 'application/vnd.github+json',
	'User-Agent': user
}
)

response.text
```
